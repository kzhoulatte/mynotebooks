```python
import numpy as np
import seaborn as sns; 
sns.set(rc={'figure.figsize':(12.7,9.27)})
sns.set_style("white")


N = 100

x = np.linspace(0,2*np.pi,N)
y = np.random.random(N)*np.sin(x) + np.random.random(N) * 0.8
for i in range(int(N*3/4),int(N*3/4)+10):
    y[i] = -1

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.figure()
plt.plot(x, y,'o')
plt.xticks([])
plt.yticks([])
plt.figure()
plt.plot(x, smooth(y,3), 'r-', lw=5)
plt.xticks([])
plt.yticks([])
plt.figure()
plt.plot(x, smooth(y,19), 'g-', lw=5)
# plt.xlabel('Timestamp', fontsize=16)
plt.xticks([])
plt.yticks([])
```

```python
xx = x[:N-10]
yy = y[:N-10]

for i in range(int(N*3/4)):
    yy[i] = y[i]

for i in range(int(N*3/4),N-10):
    yy[i] = y[i+10]

f = interp1d(xx, yy)
f2 = interp1d(xx, yy, kind='cubic')   
    
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
plt.figure()
plt.plot(xx, yy,'o')
plt.xticks([])
plt.yticks([])

plt.figure()
plt.plot(xx, f(xx), 'r-', lw=5)
plt.xticks([])
plt.yticks([])

plt.figure()
plt.plot(xx, f2(xx), 'r-', lw=5)
plt.xticks([])
plt.yticks([])


plt.figure()
plt.plot(xx, yy, 'r-', lw=5)
plt.xticks([])
plt.yticks([])
plt.figure()
plt.plot(xx, smooth(yy,9), 'g-', lw=5)
# plt.xlabel('Timestamp', fontsize=16)
plt.xticks([])
plt.yticks([])

plt.figure()
plt.plot(xx, yy, 'o', xx, f(xx), '-', xx, f2(xx), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.xticks([])
plt.yticks([])
plt.show()
```
